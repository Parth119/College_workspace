```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm

file = ("download.jfif")
image1 = imread(file,as_gray=True)

image3 = image1[5:1000,5:100]
image2 = resize(image3,(30,30),mode='edge')
plt.imshow(image3,cmap=cm.gray)
plt.show()
print("Data Type: %s ,shape: %s"%(type(image2),image2.shape))
```

```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm

file = ("download.jfif")
image1 = imread(file,as_gray=True)

image3 = image1[5:1000,5:100]
image2 = resize(image3,(30,30),mode='edge')
image_row = image2.flatten()

plt.show()
print("Data Type: %s ,shape: %s"%(type(image_row),image_row.shape))
```

```python
from lxml import objectify
import pandas as pd
xml = objectify.parse(open('XMLData.xml'))
root = xml.getroot()
df= pd.DataFrame(columns=('Number','String','Boolean'))

for i in range(4):
    obj= root.getchildren()[i].getchildren()
    row= dict(zip(['Number','String','Boolean'],[obj[0].text,obj[1].text,obj[2].text]))
    row_s= pd.Series(row)
    row_s.name=i
    df=df.append(row_s)
print(df)
```
